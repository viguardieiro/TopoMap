```python
import pandas as pd
from TopoMap import TopoMap
import matplotlib.pyplot as plt
```

```python
def plot_data_projection(df, data, proj):
    colors = ['blue', 'red', 'green']

    fig = plt.figure(figsize=(8,12))


    ax = fig.add_subplot(3, 2, 1, projection='3d')
    for cluster in df['ClusterId'].unique():
        i_cluster = list(df[df['ClusterId']==cluster].index)
        ax.scatter(data[i_cluster,0], data[i_cluster,1], data[i_cluster,2],
                c=colors[cluster], s=5)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_zticks([])
    ax.set_title('Original points')

    ax = fig.add_subplot(3, 2, 2)
    for cluster in df['ClusterId'].unique():
        i_cluster = list(df[df['ClusterId']==cluster].index)
        ax.scatter(proj[i_cluster,0], proj[i_cluster,1],
                c=colors[cluster], s=5)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_title('TopoMap projection')

    fig.tight_layout()

    return fig
```

```python
df_blobs = pd.read_csv('data/3blobs.csv')
data_blobs = df_blobs[['x','y','z']].values

topomap_blobs = TopoMap(data_blobs)
proj_blobs = topomap_blobs.run()
```

```python
fig_blobs = plot_data_projection(df_blobs, data_blobs, proj_blobs)
fig_blobs.suptitle('3 Blobs Dataset')
fig_blobs.tight_layout()
plt.show()
```

```python
df_rings = pd.read_csv('data/3rings.csv')
data_rings = df_rings[['x','y','z']].values

topomap_rings = TopoMap(data_rings)
proj_rings = topomap_rings.run()
```

```python
fig_rings = plot_data_projection(df_rings, data_rings, proj_rings)
fig_rings.suptitle('3 Rings Dataset')
fig_rings.tight_layout()
plt.show()
```
